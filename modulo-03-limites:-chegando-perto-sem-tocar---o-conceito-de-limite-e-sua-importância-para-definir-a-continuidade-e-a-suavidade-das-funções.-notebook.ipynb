## ğŸ¤” TÃ¡, mas o que Ã© um Limite Afinal?

Um **limite** Ã© o valor que uma funÃ§Ã£o se aproxima quando a variÃ¡vel independente se aproxima de um determinado ponto.

Matematicamente escrevemos:

$$\lim_{x \to a} f(x) = L$$

Isso significa: "O limite de f(x) quando x se aproxima de a Ã© igual a L"

**Analogia do Crush**: Ã‰ como quando vocÃª quer chegar perto do seu crush na escola. VocÃª vai se aproximando: da mesa do fundo, pra mesa do meio, pra mesa da frente... VocÃª tÃ¡ chegando cada vez mais perto, mas pode ser que nunca chegue exatamente onde ele tÃ¡ sentado! ğŸ˜…

### Tipos de AproximaÃ§Ã£o:

1. **Pela esquerda**: $\lim_{x \to a^-} f(x)$ (chegando por valores menores que a)
2. **Pela direita**: $\lim_{x \to a^+} f(x)$ (chegando por valores maiores que a)
3. **Bilateral**: quando os dois limites acima sÃ£o iguais

**Dica do Pedro**: O limite existe quando conseguimos chegar no mesmo "destino" independente do "caminho" que escolhemos!

```python
# Vamos criar nossa primeira visualizaÃ§Ã£o de limite!
def visualizar_limite_basico():
    x = np.linspace(-3, 5, 1000)
    
    # FunÃ§Ã£o simples: f(x) = x^2
    y = x**2
    
    fig, ax = plt.subplots(figsize=(10, 6))
    
    # Plotar a funÃ§Ã£o
    ax.plot(x, y, 'b-', linewidth=2, label='f(x) = xÂ²')
    
    # Ponto de interesse (x = 2)
    ponto_x = 2
    ponto_y = ponto_x**2
    
    # Marcar o ponto
    ax.plot(ponto_x, ponto_y, 'ro', markersize=10, label=f'Ponto (2, 4)')
    
    # Valores se aproximando de 2
    aproximacoes = [1.5, 1.8, 1.9, 1.99, 2.01, 2.1, 2.2, 2.5]
    for aprox in aproximacoes:
        y_aprox = aprox**2
        ax.plot(aprox, y_aprox, 'go', markersize=6, alpha=0.7)
        ax.plot([aprox, aprox], [0, y_aprox], 'g--', alpha=0.5)
    
    ax.set_xlabel('x')
    ax.set_ylabel('f(x)')
    ax.set_title('Limite: Aproximando-se do ponto (2, 4)')
    ax.grid(True, alpha=0.3)
    ax.legend()
    
    plt.tight_layout()
    plt.show()
    
    print("ğŸ¯ Repare como os pontos verdes se aproximam do ponto vermelho!")
    print("ğŸ’¡ Quando x se aproxima de 2, f(x) se aproxima de 4")

visualizar_limite_basico()
```

## ğŸ§® Calculando Limites na PrÃ¡tica

Existem vÃ¡rias tÃ©cnicas para calcular limites. Vamos ver as principais:

### 1. SubstituiÃ§Ã£o Direta
Se a funÃ§Ã£o Ã© contÃ­nua no ponto, sÃ³ substituir:

$$\lim_{x \to 2} x^2 = 2^2 = 4$$

### 2. FatoraÃ§Ã£o (quando dÃ¡ 0/0)
$$\lim_{x \to 2} \frac{x^2-4}{x-2} = \lim_{x \to 2} \frac{(x-2)(x+2)}{x-2} = \lim_{x \to 2} (x+2) = 4$$

### 3. Limites no Infinito
$$\lim_{x \to \infty} \frac{1}{x} = 0$$

**Dica do Pedro**: Quando der 0/0 ou âˆ/âˆ, nÃ£o desespere! Ã‰ sÃ³ uma "forma indeterminada". Tem sempre um jeito de resolver! ğŸ¤“

```python
# Vamos calcular alguns limites usando SymPy
x = symbols('x')

print("ğŸ§® Calculando Limites com SymPy")
print("=" * 40)

# Limite simples por substituiÃ§Ã£o
limite1 = limit(x**2, x, 2)
print(f"1. lim(xÂ²) quando xâ†’2 = {limite1}")

# Limite com forma indeterminada 0/0
limite2 = limit((x**2 - 4)/(x - 2), x, 2)
print(f"2. lim((xÂ²-4)/(x-2)) quando xâ†’2 = {limite2}")

# Limite no infinito
limite3 = limit(1/x, x, oo)
print(f"3. lim(1/x) quando xâ†’âˆ = {limite3}")

# Limite trigonomÃ©trico famoso
limite4 = limit(sin(x)/x, x, 0)
print(f"4. lim(sen(x)/x) quando xâ†’0 = {limite4}")

# Limite exponencial
limite5 = limit((1 + 1/x)**x, x, oo)
print(f"5. lim((1+1/x)^x) quando xâ†’âˆ = {limite5}")

print("\nğŸ’¡ O Ãºltimo Ã© o famoso nÃºmero e (Euler)!")
```

## ğŸ“Š Visualizando Limites ProblemÃ¡ticos

Nem sempre o limite existe ou Ã© Ã³bvio. Vamos ver alguns casos interessantes:

### Caso 1: Descontinuidade RemovÃ­vel
Quando temos um "buraco" na funÃ§Ã£o, mas o limite existe.

### Caso 2: Descontinuidade de Salto
Quando a funÃ§Ã£o "pula" de um valor para outro.

### Caso 3: Descontinuidade Infinita
Quando a funÃ§Ã£o "explode" para o infinito.

![](/Users/pedroguth/Downloads/Projetos/Book Maker/5-Imagens/cÃ¡lculo-para-ia-modulo-03_img_02.png)

```python
# Visualizando diferentes tipos de descontinuidade
def visualizar_descontinuidades():
    fig, axes = plt.subplots(1, 3, figsize=(15, 5))
    
    # Caso 1: Descontinuidade RemovÃ­vel (buraco)
    x1 = np.linspace(-3, 5, 1000)
    x1 = x1[x1 != 2]  # Remove o ponto x=2
    y1 = (x1**2 - 4) / (x1 - 2)  # Simplifica para x+2, mas com buraco em x=2
    
    axes[0].plot(x1, y1, 'b-', linewidth=2)
    axes[0].plot(2, 4, 'wo', markersize=8, markeredgecolor='red', markeredgewidth=2)  # Buraco
    axes[0].plot(2, 4, 'ro', markersize=4)  # Onde deveria estar
    axes[0].set_title('Descontinuidade RemovÃ­vel\n(Buraco)')
    axes[0].set_xlabel('x')
    axes[0].set_ylabel('f(x)')
    axes[0].grid(True, alpha=0.3)
    axes[0].set_ylim(0, 8)
    
    # Caso 2: Descontinuidade de Salto
    x2 = np.linspace(-2, 4, 1000)
    y2 = np.where(x2 < 1, x2**2, x2**2 + 2)  # FunÃ§Ã£o com salto em x=1
    
    # Plotar as duas partes separadamente para mostrar o salto
    x2_esq = x2[x2 < 1]
    y2_esq = x2_esq**2
    x2_dir = x2[x2 >= 1]
    y2_dir = x2_dir**2 + 2
    
    axes[1].plot(x2_esq, y2_esq, 'b-', linewidth=2)
    axes[1].plot(x2_dir, y2_dir, 'b-', linewidth=2)
    axes[1].plot(1, 1, 'ro', markersize=6)  # Ponto Ã  esquerda
    axes[1].plot(1, 3, 'bo', markersize=6)  # Ponto Ã  direita
    axes[1].set_title('Descontinuidade de Salto')
    axes[1].set_xlabel('x')
    axes[1].set_ylabel('f(x)')
    axes[1].grid(True, alpha=0.3)
    
    # Caso 3: Descontinuidade Infinita (assÃ­ntota)
    x3_pos = np.linspace(0.01, 3, 500)
    x3_neg = np.linspace(-3, -0.01, 500)
    y3_pos = 1/x3_pos
    y3_neg = 1/x3_neg
    
    axes[2].plot(x3_pos, y3_pos, 'b-', linewidth=2)
    axes[2].plot(x3_neg, y3_neg, 'b-', linewidth=2)
    axes[2].axvline(x=0, color='red', linestyle='--', alpha=0.7, label='AssÃ­ntota')
    axes[2].set_title('Descontinuidade Infinita\n(AssÃ­ntota)')
    axes[2].set_xlabel('x')
    axes[2].set_ylabel('f(x)')
    axes[2].set_ylim(-10, 10)
    axes[2].grid(True, alpha=0.3)
    axes[2].legend()
    
    plt.tight_layout()
    plt.show()
    
    print("ğŸ¯ TrÃªs tipos de 'problemas' que uma funÃ§Ã£o pode ter:")
    print("1ï¸âƒ£ Buraco: O limite existe, mas a funÃ§Ã£o nÃ£o estÃ¡ definida no ponto")
    print("2ï¸âƒ£ Salto: Os limites laterais sÃ£o diferentes")
    print("3ï¸âƒ£ AssÃ­ntota: A funÃ§Ã£o 'explode' para infinito")

visualizar_descontinuidades()
```

## ğŸ”„ Continuidade: Quando Tudo Flui Perfeitamente

Uma funÃ§Ã£o Ã© **contÃ­nua** em um ponto quando trÃªs condiÃ§Ãµes sÃ£o satisfeitas:

1. $f(a)$ existe (a funÃ§Ã£o estÃ¡ definida no ponto)
2. $\lim_{x \to a} f(x)$ existe (o limite existe)
3. $\lim_{x \to a} f(x) = f(a)$ (o limite Ã© igual ao valor da funÃ§Ã£o)

**Analogia do Desenho**: Imagine que vocÃª tÃ¡ desenhando uma funÃ§Ã£o. Se conseguir desenhar sem tirar o lÃ¡pis do papel, ela Ã© contÃ­nua! Se tiver que "pular" algum ponto, aÃ­ tem descontinuidade.

### Por que isso Ã© importante para IA?

- **FunÃ§Ãµes de ativaÃ§Ã£o** precisam ser contÃ­nuas para o backpropagation funcionar
- **FunÃ§Ãµes de perda** contÃ­nuas garantem otimizaÃ§Ã£o suave
- **Gradiente descendente** sÃ³ funciona bem com funÃ§Ãµes "comportadas"

**Dica do Pedro**: Na IA, continuidade = suavidade = gradientes bem definidos = algoritmo feliz! ğŸ˜Š

![](/Users/pedroguth/Downloads/Projetos/Book Maker/5-Imagens/cÃ¡lculo-para-ia-modulo-03_img_03.png)

```python
# Testando continuidade de diferentes funÃ§Ãµes
def testar_continuidade(func_sympy, ponto, nome_funcao):
    """Testa se uma funÃ§Ã£o Ã© contÃ­nua em um ponto"""
    try:
        # CondiÃ§Ã£o 1: f(a) existe?
        valor_funcao = func_sympy.subs(x, ponto)
        
        # CondiÃ§Ã£o 2: limite existe?
        limite_bilateral = limit(func_sympy, x, ponto)
        
        # CondiÃ§Ã£o 3: limite = f(a)?
        continua = (valor_funcao == limite_bilateral)
        
        print(f"\nğŸ“Š Testando {nome_funcao} no ponto x = {ponto}")
        print(f"   f({ponto}) = {valor_funcao}")
        print(f"   lim f(x) quando xâ†’{ponto} = {limite_bilateral}")
        print(f"   ContÃ­nua? {'âœ… SIM' if continua else 'âŒ NÃƒO'}")
        
        return continua
        
    except:
        print(f"\nğŸ“Š Testando {nome_funcao} no ponto x = {ponto}")
        print(f"   âŒ FunÃ§Ã£o nÃ£o estÃ¡ definida no ponto ou limite nÃ£o existe")
        return False

print("ğŸ” TESTE DE CONTINUIDADE")
print("=" * 50)

# FunÃ§Ã£o contÃ­nua
f1 = x**2
testar_continuidade(f1, 2, "f(x) = xÂ²")

# FunÃ§Ã£o com descontinuidade removÃ­vel
f2 = (x**2 - 4)/(x - 2)
testar_continuidade(f2, 2, "f(x) = (xÂ²-4)/(x-2)")

# FunÃ§Ã£o contÃ­nua (polinÃ´mio)
f3 = x**3 - 2*x + 1
testar_continuidade(f3, 1, "f(x) = xÂ³ - 2x + 1")

# FunÃ§Ã£o trigonomÃ©trica contÃ­nua
f4 = sin(x)
testar_continuidade(f4, 0, "f(x) = sen(x)")
```

## ğŸ¯ AplicaÃ§Ã£o em IA: FunÃ§Ãµes de AtivaÃ§Ã£o

Vamos ver como os limites se aplicam Ã s funÃ§Ãµes de ativaÃ§Ã£o que usamos em redes neurais!

### Principais FunÃ§Ãµes de AtivaÃ§Ã£o:

1. **Sigmoid**: $\sigma(x) = \frac{1}{1+e^{-x}}$
2. **Tanh**: $\tanh(x) = \frac{e^x - e^{-x}}{e^x + e^{-x}}$
3. **ReLU**: $\text{ReLU}(x) = \max(0, x)$
4. **Leaky ReLU**: $\text{LeakyReLU}(x) = \max(0.01x, x)$

**Pergunta importante**: Todas essas funÃ§Ãµes sÃ£o contÃ­nuas? ğŸ¤”

![](/Users/pedroguth/Downloads/Projetos/Book Maker/5-Imagens/cÃ¡lculo-para-ia-modulo-03_img_04.png)

```python
# Implementando e visualizando funÃ§Ãµes de ativaÃ§Ã£o
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def tanh_func(x):
    return np.tanh(x)

def relu(x):
    return np.maximum(0, x)

def leaky_relu(x, alpha=0.01):
    return np.where(x > 0, x, alpha * x)

# Visualizando as funÃ§Ãµes de ativaÃ§Ã£o
x_vals = np.linspace(-5, 5, 1000)

fig, axes = plt.subplots(2, 2, figsize=(15, 10))

# Sigmoid
axes[0,0].plot(x_vals, sigmoid(x_vals), 'b-', linewidth=2)
axes[0,0].set_title('Sigmoid: Ïƒ(x) = 1/(1+e^(-x))')
axes[0,0].grid(True, alpha=0.3)
axes[0,0].axhline(y=0.5, color='red', linestyle='--', alpha=0.5)
axes[0,0].axvline(x=0, color='red', linestyle='--', alpha=0.5)

# Tanh
axes[0,1].plot(x_vals, tanh_func(x_vals), 'g-', linewidth=2)
axes[0,1].set_title('Tanh: (e^x - e^(-x))/(e^x + e^(-x))')
axes[0,1].grid(True, alpha=0.3)
axes[0,1].axhline(y=0, color='red', linestyle='--', alpha=0.5)
axes[0,1].axvline(x=0, color='red', linestyle='--', alpha=0.5)

# ReLU
axes[1,0].plot(x_vals, relu(x_vals), 'r-', linewidth=2)
axes[1,0].set_title('ReLU: max(0, x)')
axes[1,0].grid(True, alpha=0.3)
axes[1,0].axhline(y=0, color='red', linestyle='--', alpha=0.5)
axes[1,0].axvline(x=0, color='red', linestyle='--', alpha=0.5)

# Leaky ReLU
axes[1,1].plot(x_vals, leaky_relu(x_vals), 'm-', linewidth=2)
axes[1,1].set_title('Leaky ReLU: max(0.01x, x)')
axes[1,1].grid(True, alpha=0.3)
axes[1,1].axhline(y=0, color='red', linestyle='--', alpha=0.5)
axes[1,1].axvline(x=0, color='red', linestyle='--', alpha=0.5)

for ax in axes.flat:
    ax.set_xlabel('x')
    ax.set_ylabel('f(x)')

plt.tight_layout()
plt.show()

print("ğŸ¯ AnÃ¡lise de Continuidade das FunÃ§Ãµes de AtivaÃ§Ã£o:")
print("âœ… Sigmoid: ContÃ­nua em todos os pontos")
print("âœ… Tanh: ContÃ­nua em todos os pontos")
print("âš ï¸  ReLU: ContÃ­nua, mas nÃ£o diferenciÃ¡vel em x=0")
print("âš ï¸  Leaky ReLU: ContÃ­nua, mas nÃ£o diferenciÃ¡vel em x=0")
print("\nğŸ’¡ A continuidade garante que o gradiente descendente funcione!")
```

## ğŸ“ˆ Limites e Comportamento AssintÃ³tico

Na IA, frequentemente nos interessamos pelo comportamento das funÃ§Ãµes quando $x \to \infty$ ou $x \to -\infty$.

### Por que isso importa?

- **Gradiente Vanishing**: Quando gradientes ficam muito pequenos
- **Gradiente Exploding**: Quando gradientes ficam muito grandes
- **SaturaÃ§Ã£o de neurÃ´nios**: Quando ativaÃ§Ãµes ficam muito extremas

Vamos calcular alguns limites importantes:

$$\lim_{x \to \infty} \frac{1}{1+e^{-x}} = 1 \quad \text{(Sigmoid satura)}$$

$$\lim_{x \to -\infty} \frac{1}{1+e^{-x}} = 0 \quad \text{(Sigmoid satura)}$$

$$\lim_{x \to \infty} \max(0,x) = \infty \quad \text{(ReLU nÃ£o satura)}$$

```python
# Analisando comportamento assintÃ³tico das funÃ§Ãµes de ativaÃ§Ã£o
print("ğŸ”„ ANÃLISE DE COMPORTAMENTO ASSINTÃ“TICO")
print("=" * 50)

# Definindo as funÃ§Ãµes simbolicamente
sigmoid_symb = 1/(1 + exp(-x))
tanh_symb = (exp(x) - exp(-x))/(exp(x) + exp(-x))

# Calculando limites no infinito
print("ğŸ“Š Limites quando x â†’ +âˆ:")
limite_sigmoid_pos = limit(sigmoid_symb, x, oo)
print(f"   Sigmoid: {limite_sigmoid_pos}")

limite_tanh_pos = limit(tanh_symb, x, oo)
print(f"   Tanh: {limite_tanh_pos}")

print(f"   ReLU: +âˆ (nÃ£o satura)")
print(f"   Leaky ReLU: +âˆ (nÃ£o satura)")

print("\nğŸ“Š Limites quando x â†’ -âˆ:")
limite_sigmoid_neg = limit(sigmoid_symb, x, -oo)
print(f"   Sigmoid: {limite_sigmoid_neg}")

limite_tanh_neg = limit(tanh_symb, x, -oo)
print(f"   Tanh: {limite_tanh_neg}")

print(f"   ReLU: 0 (mata neurÃ´nios mortos)")
print(f"   Leaky ReLU: -âˆ (com inclinaÃ§Ã£o pequena)")

# Visualizando saturaÃ§Ã£o
x_extreme = np.linspace(-10, 10, 1000)

fig, ax = plt.subplots(figsize=(12, 6))

ax.plot(x_extreme, sigmoid(x_extreme), 'b-', linewidth=2, label='Sigmoid')
ax.plot(x_extreme, tanh_func(x_extreme), 'g-', linewidth=2, label='Tanh')
ax.plot(x_extreme, np.clip(relu(x_extreme), 0, 10), 'r-', linewidth=2, label='ReLU (cortada em 10)')

# Linhas de saturaÃ§Ã£o
ax.axhline(y=1, color='blue', linestyle='--', alpha=0.5, label='SaturaÃ§Ã£o Sigmoid')
ax.axhline(y=0, color='blue', linestyle='--', alpha=0.5)
ax.axhline(y=1, color='green', linestyle='--', alpha=0.3)
ax.axhline(y=-1, color='green', linestyle='--', alpha=0.3, label='SaturaÃ§Ã£o Tanh')

ax.set_xlabel('x')
ax.set_ylabel('f(x)')
ax.set_title('Comportamento AssintÃ³tico das FunÃ§Ãµes de AtivaÃ§Ã£o')
ax.grid(True, alpha=0.3)
ax.legend()
ax.set_ylim(-2, 5)

plt.tight_layout()
plt.show()

print("\nğŸ’¡ Dica do Pedro: Sigmoid e Tanh saturam (problema do gradiente que desaparece)")
print("ğŸš€ ReLU nÃ£o satura para valores positivos (por isso Ã© tÃ£o popular!)")
```

Considere a funÃ§Ã£o:
$$f(x) = \begin{cases}

Ela Ã© contÃ­nua em x = 2?

```python
# EXERCÃCIO 2: Analisando continuidade de funÃ§Ã£o por partes
def analisar_continuidade_partes():
    print("ğŸ® EXERCÃCIO 2: FunÃ§Ã£o por Partes")
    print("=" * 40)
    
    # Definindo a funÃ§Ã£o por partes
    def f_partes(x_val):
        if x_val < 2:
            return x_val**2
        elif x_val == 2:
            return 4
        else:
            return 2*x_val
    
    # Testando continuidade em x = 2
    ponto = 2
    
    # 1. f(2) existe?
    valor_funcao = f_partes(2)
    print(f"f(2) = {valor_funcao}")
    
    # 2. Limite pela esquerda
    limite_esquerda = limit(x**2, x, 2)
    print(f"lim f(x) quando xâ†’2â» = {limite_esquerda}")
    
    # 3. Limite pela direita
    limite_direita = limit(2*x, x, 2)
    print(f"lim f(x) quando xâ†’2âº = {limite_direita}")
    
    # 4. AnÃ¡lise
    limite_bilateral_existe = (limite_esquerda == limite_direita)
    continua = limite_bilateral_existe and (limite_esquerda == valor_funcao)
    
    print(f"\nğŸ“Š AnÃ¡lise:")
    print(f"   Limite bilateral existe? {'âœ… SIM' if limite_bilateral_existe else 'âŒ NÃƒO'}")
    print(f"   Limite = f(2)? {'âœ… SIM' if (limite_esquerda == valor_funcao) else 'âŒ NÃƒO'}")
    print(f"   FunÃ§Ã£o contÃ­nua em x=2? {'âœ… SIM' if continua else 'âŒ NÃƒO'}")
    
    # Visualizando
    x_vals = np.linspace(0, 4, 1000)
    y_vals = []
    
    for x_val in x_vals:
        if x_val < 2:
            y_vals.append(x_val**2)
        else:
            y_vals.append(2*x_val)
    
    plt.figure(figsize=(10, 6))
    
    # Plotar partes da funÃ§Ã£o
    x_esq = x_vals[x_vals < 2]
    y_esq = x_esq**2
    x_dir = x_vals[x_vals >= 2]
    y_dir = 2*x_dir
    
    plt.plot(x_esq, y_esq, 'b-', linewidth=2, label='f(x) = xÂ² (x < 2)')
    plt.plot(x_dir, y_dir, 'g-', linewidth=2, label='f(x) = 2x (x > 2)')
    
    # Marcar pontos importantes
    plt.plot(2, 4, 'ro', markersize=8, label='f(2) = 4')
    plt.plot(2, 4, 'bo', markersize=6, alpha=0.7, label='lim esquerda = 4')
    plt.plot(2, 4, 'go', markersize=4, alpha=0.7, label='lim direita = 4')
    
    plt.axvline(x=2, color='red', linestyle='--', alpha=0.5)
    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.title('FunÃ§Ã£o por Partes - AnÃ¡lise de Continuidade')
    plt.grid(True, alpha=0.3)
    plt.legend()
    plt.tight_layout()
    plt.show()
    
    return continua

resultado = analisar_continuidade_partes()
print(f"\nğŸ¯ Resultado final: A funÃ§Ã£o {'Ã‰' if resultado else 'NÃƒO Ã‰'} contÃ­nua em x = 2")
```

## ğŸ”— Conectando com os PrÃ³ximos MÃ³dulos

Agora que dominamos limites, vamos ver como eles se conectam com o que vem pela frente:

### PrÃ³ximo MÃ³dulo: Derivadas
A derivada Ã© definida como um **limite especial**:

$$f'(x) = \lim_{h \to 0} \frac{f(x+h) - f(x)}{h}$$

### PreparaÃ§Ã£o para Gradiente Descendente
- FunÃ§Ãµes contÃ­nuas â†’ Gradientes bem definidos
- Limites no infinito â†’ Comportamento de saturaÃ§Ã£o
- Continuidade â†’ Algoritmos de otimizaÃ§Ã£o funcionam

**Dica do Pedro**: Pense nos limites como a "cola" que conecta pontos discretos em curvas suaves. Na IA, isso Ã© essencial para que nossos algoritmos "naveguem" suavemente pelo espaÃ§o de parÃ¢metros! ğŸŒŠ

```python
# Preparando para o prÃ³ximo mÃ³dulo: Derivadas como Limites
print("ğŸ”— PREPARAÃ‡ÃƒO PARA DERIVADAS")
print("=" * 40)

def aproximar_derivada(func, ponto, h_values):
    """Aproxima a derivada usando a definiÃ§Ã£o por limite"""
    derivadas_aprox = []
    
    for h in h_values:
        # (f(x+h) - f(x)) / h
        slope = (func(ponto + h) - func(ponto)) / h
        derivadas_aprox.append(slope)
    
    return derivadas_aprox

# Testando com f(x) = xÂ²
def f_quadratica(x):
    return x**2

ponto_teste = 3
h_vals = [1, 0.1, 0.01, 0.001, 0.0001, 0.00001]

aproximacoes = aproximar_derivada(f_quadratica, ponto_teste, h_vals)

print(f"Aproximando f'(3) para f(x) = xÂ²:")
print(f"(Resposta exata: 6)\n")

for i, (h, aprox) in enumerate(zip(h_vals, aproximacoes)):
    erro = abs(aprox - 6)
    print(f"h = {h:8} â†’ f'(3) â‰ˆ {aprox:8.5f} (erro: {erro:.5f})")

# Visualizando a convergÃªncia
plt.figure(figsize=(10, 6))
plt.semilogx(h_vals, aproximacoes, 'bo-', linewidth=2, markersize=8, label='AproximaÃ§Ãµes')
plt.axhline(y=6, color='red', linestyle='--', linewidth=2, label='Derivada exata = 6')
plt.xlabel('h (tamanho do passo)')
plt.ylabel('AproximaÃ§Ã£o da derivada')
plt.title('ConvergÃªncia da DefiniÃ§Ã£o de Derivada por Limite')
plt.grid(True, alpha=0.3)
plt.legend()
plt.gca().invert_xaxis()  # h diminuindo da esquerda para direita
plt.tight_layout()
plt.show()

print("\nğŸ¯ Liiindo! Conforme h â†’ 0, nossa aproximaÃ§Ã£o â†’ derivada exata!")
print("ğŸš€ No prÃ³ximo mÃ³dulo vamos formalizar isso e aprender as regras!")
```
